```python
import copy

a = [[1, 2, 3], [4, 5, 6]]

# Shallow copy
b = copy.copy(a)
b[0][0] = 999
print(a)  # [[999, 2, 3], [4, 5, 6]] - CHANGED! ❌
print(b)   # [[999, 2, 3], [4, 5, 6]]
```

```python
# Deep copy
a = [[1, 2, 3], [4, 5, 6]]
b = copy.deepcopy(a)
b[0][0] = 999
print(a)  # [[1, 2, 3], [4, 5, 6]] - unchanged ✓
print(b)      # [[999, 2, 3], [4, 5, 6]]
```
